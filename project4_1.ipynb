```python
import numpy
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ['AAPL', 'FB', 'MSFT','V', 'DIS', 'NVDA']
stock_data = []


for t in tickers:
  stock_data.append(yf.download(t, '2020-11-19', '2021-11-19'))
```

```python
df = pd.DataFrame(columns=['Date', 'Low', 'High', 'Close', 'Open'])

for i in  stock_data:
  print(i)
```

```python
for index, val in enumerate(stock_data):
  val['Adj Close'].plot(label=tickers[index], figsize=(15,7))
plt.title('Stock Prices of ' + ', '.join(tickers))
plt.legend()
```
